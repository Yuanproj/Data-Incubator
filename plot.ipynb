```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../data/kaiko-trades/gz_v1/Binance/BTCUSDT/2019_07/Binance_BTCUSDT_trades_2019_07_19.csv.gz')
```

```python
data['date'] = pd.to_datetime(data['date'], unit='ms')
```

```python
data.head()
```

```python
data.groupby(pd.cut(data['date'], 20, labels=range(20)))[['price']].mean().plot()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,8))
sns.catplot(ax=ax, x='date', y='amount', hue='sell', kind='bar', data=data.groupby([pd.cut(data['date'], 15, labels=range(15)), 'sell'])['amount'].mean().reset_index())
```
